Question: Why does Matplotlib treat images as NumPy arrays?

Answer:Because an image can be represented as a matrix of pixel values, and NumPy arrays provide an efficient way to store and manipulate numerical data.


```python
from skimage import io, img_as_float
import matplotlib.pyplot as plt


img_sk = io.imread("/content/drive/MyDrive/tobias.jpg")

img_float = img_as_float(img_sk)

print("Min pixel value:", img_float.min())
print("Max pixel value:", img_float.max())

plt.imshow(img_float)
plt.axis("off")
```

Question: Why is using astype(float) discouraged compared to img_as_float? Explain briefly.

Answer: Because astype(float) only changes the data type without rescaling the pixel values, while img_as_float properly converts and normalizes the image intensities to the correct range.

```python
import cv2
import matplotlib.pyplot as plt

img_gray = cv2.imread("/content/drive/MyDrive/tobias.jpg", cv2.IMREAD_GRAYSCALE)

img_bgr = cv2.imread("/content/drive/MyDrive/tobias.jpg", cv2.IMREAD_COLOR)

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
edges = cv2.Canny(img_gray, threshold1=100, threshold2=200)

plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.show()
```

Answer: The difference lies in the color channel order: RGB arranges colors as Red, Green, Blue, while BGR uses Blue, Green, Red. OpenCV uses the BGR format by default, which may result in incorrect colors unless the image is converted.
